```python
w_source = water["country_name"].tolist()
w_source[:10]
```

```python
d_w_source = {}
for w in w_source:
  if w is np.nan:
    d_w_source["null"] =  d_w_source.get("null", 0) + 1
  else:
    d_w_source[w] = d_w_source.get(w, 0) + 1
```

```python
lc = [ (country, d_w_source[country]) for country in d_w_source.keys()]
```

```python
swap = lambda x: (x[1],x[0])
l_sorted_source_number = list(map(swap, sorted(map(swap, lc), reverse=True)))
```
